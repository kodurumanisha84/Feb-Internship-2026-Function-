```python
#Problem Statement 1:Smart Parking Lot Management System
def smart_parking(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

        # Track peak parking usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    # Parking Status Check
    if current_vehicles > capacity:
        status = "Parking Full! Alert!"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

```python
#Problem Statement 2:Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):

    # Base time calculation
    delivery_time = distance * 5

    # Traffic conditions
    if traffic == "Low":
        delivery_time += 5
    elif traffic == "Medium":
        delivery_time += 10
    elif traffic == "High":
        delivery_time += 15

    # Weather conditions
    if weather == "Cloudy":
        delivery_time += 5
    elif weather == "Rainy":
        delivery_time += 10

    print("Estimated Delivery Time:", delivery_time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
#Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
        suggestion = "Open additional show immediately!"
    elif occupancy >= 70:
        status = "Almost Full"
        suggestion = "Consider opening extra shows."
    else:
        status = "Seats Available"
        suggestion = "No additional show required."

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

seat_occupancy(total_seats, booked_seats)
```

```python
# Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):

    # Calculate average CPU usage
    average_cpu = sum(cpu_readings) / len(cpu_readings)

    # Classify server status
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(average_cpu), "%")
    print("Server Status:", status)


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
#Problem Statement 5:Smart Classroom Resource Usage Monitor
def resource_monitor(resource_usage):

    overused_resources = []

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)

    # Display results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(resource_usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller
def event_registration(capacity, registrations):

    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
capacity = 100
registrations = 105

event_registration(capacity, registrations)
```
